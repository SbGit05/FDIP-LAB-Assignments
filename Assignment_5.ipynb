```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Path to image
image_path = r"D:\shreedhar\Movies\sample2\example.jpg"

# Read image in grayscale (since histogram equalization is usually done on single channel)
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if img is None:
    raise FileNotFoundError("Image not found. Please check the path again!")

# --- Contrast Stretching ---
# Get minimum and maximum pixel intensity
min_val = np.min(img)
max_val = np.max(img)

# Apply contrast stretching formula
stretched = ((img - min_val) / (max_val - min_val) * 255).astype(np.uint8)

# --- Histogram Equalization ---
equalized = cv2.equalizeHist(img)

# --- Plotting ---
titles = ['Original Image', 'Contrast Stretched', 'Histogram Equalized']
images = [img, stretched, equalized]

plt.figure(figsize=(12, 8))

for i in range(3):
    # Show image
    plt.subplot(3, 2, 2*i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

    # Show histogram
    plt.subplot(3, 2, 2*i+2)
    plt.hist(images[i].ravel(), bins=256, range=[0,256], color='black')
    plt.title(f"{titles[i]} Histogram")

plt.tight_layout()
plt.show()
```
